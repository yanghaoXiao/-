```python
import numpy as np
import sympy
```

```python
w1 = np.array([[0,0],[2,0],[2,2],[0,2]])
w2 = np.array([[4,4],[6,4],[6,6],[4,6]])
```

```python
m1 = np.mean(w1,axis=0)
m2 = np.mean(w2,axis=0)
c1 = np.cov(w1,rowvar=False,bias=True)
c2 = np.cov(w2,rowvar=False,bias=True)
print(c1 == c2)
```

```python
c = c1
inv_c = np.linalg.inv(c)
x1 = sympy.symbols("x1")
x2 = sympy.symbols("x2")
x = np.array([x1,x2]) 
```

```python
# 由计算可知协方差矩阵c1=c2，此时两类问题且正态分布的判别界面为x的线性函数，不妨设为ax+b=0,其中a和x为二维向量
a = np.dot((m1-m2).T,inv_c)
b = -1/2 * np.dot(np.dot(m1.T,inv_c),m1) + 1/2 * np.dot(np.dot(m2.T,inv_c),m2)
d = np.dot(a,x)+b
print("判别界面的方程为:%s" %d +' = 0')
```
